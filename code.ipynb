```python
def get_price_by_year(year_selected, price_data, financial_year, mode):
     #General function to compute max or min price for a given year.
    filter_year = (financial_year == year_selected) & (price_data > 0)
    if mode == "max":
        return np.max(price_data[filter_year])
    elif mode == "min":
        return np.min(price_data[filter_year])
    
# Calculate max and min prices for each year
line_max = [get_price_by_year(year, max_selling_price, PriceRange_year, mode="max") for year in PriceRange_years]
line_min = [get_price_by_year(year, min_selling_price, PriceRange_year, mode="min") for year in PriceRange_years]

# Plot the data
plt.figure(figsize=(10, 6))  # Set figure size

plt.plot(PriceRange_years, line_max, marker='o', linestyle='-', color='red', label="Maximum Selling Price")
plt.plot(PriceRange_years, line_min, marker='o', linestyle='--', color='blue', label="Minimum Selling Price")

# Add gridlines, labels, and title
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xlabel("Years", fontsize=12)
plt.ylabel("Prices (SGD)", fontsize=12)
plt.title("HDB Flat Maximum and Minimum Selling Prices by Year", fontsize=14, fontweight='bold')

# Customize ticks and legend
plt.xticks(PriceRange_years, rotation=45, fontsize=10)
plt.legend(fontsize=10, loc="upper left")

# Add annotations for better readability
for x, max_price, min_price in zip(PriceRange_years, line_max, line_min):
    plt.text(x, max_price + 1000, f'{max_price:.0f}', ha='center', va='bottom', fontsize=8, color='red')
    plt.text(x, min_price - 1000, f'{min_price:.0f}', ha='center', va='top', fontsize=8, color='blue')

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

```python
# Define a function for dynamic bin calculation
def calculate_bins(data):
    # Calculate number of bins using Freedman–Diaconis rule
    q25, q75 = np.percentile(data, [25, 75])
    iqr = q75 - q25
    bin_width = 2 * iqr * len(data) ** (-1/3)
    bins = max(1, int((data.max() - data.min()) / bin_width))
    return bins

# Filter positive values
valid_max_prices = max_selling_price[max_selling_price > 0]
valid_min_prices = min_selling_price[min_selling_price > 0]

# Create subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Plot the first histogram (Maximum Selling Prices)
bins_max = calculate_bins(valid_max_prices)
axs[0].hist(valid_max_prices, bins=bins_max, color='red', edgecolor='black', alpha=0.8)
axs[0].set_title('Distribution of Maximum Selling Prices', fontsize=14, fontweight='bold')
axs[0].set_xlabel('Price', fontsize=12)
axs[0].set_ylabel('Frequency', fontsize=12)
axs[0].grid(axis='y', linestyle='--', alpha=0.6)

# Annotate mean and median
mean_max = np.mean(valid_max_prices)
median_max = np.median(valid_max_prices)
axs[0].axvline(mean_max, color='orange', linestyle='--', label=f'Mean: {mean_max:.0f}')
axs[0].axvline(median_max, color='green', linestyle='-', label=f'Median: {median_max:.0f}')
axs[0].legend()

# Plot the second histogram (Minimum Selling Prices)
bins_min = calculate_bins(valid_min_prices)
axs[1].hist(valid_min_prices, bins=bins_min, color='blue', edgecolor='black', alpha=0.8) 
axs[1].set_title('Distribution of Minimum Selling Prices', fontsize=14, fontweight='bold')
axs[1].set_xlabel('Price', fontsize=12)
axs[1].set_ylabel('Frequency', fontsize=12)
axs[1].grid(axis='y', linestyle='--', alpha=0.6)

# Annotate mean and median
mean_min = np.mean(valid_min_prices)
median_min = np.median(valid_min_prices)
axs[1].axvline(mean_min, color='orange', linestyle='--', label=f'Mean: {mean_min:.0f}')
axs[1].axvline(median_min, color='green', linestyle='-', label=f'Median: {median_min:.0f}')
axs[1].legend()

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

```python
# find the data for pie chart
def calculate_percentage(no_of_units, flat_type, selected_flat, category, selected_category):
    total_units = np.sum(no_of_units[category == selected_category])
    filter_flat_category = (flat_type == selected_flat) & (category == selected_category)
    percentage_unit = np.sum(no_of_units[filter_flat_category]) * 100 / total_units
    return percentage_unit
    
sold_percentages = np.array([calculate_percentage(no_of_units, SalesRentals_flat_type, selected_flat, category, 'Sold') 
                             for selected_flat in SalesRentals_flat_types])
rented_percentages = np.array([calculate_percentage(no_of_units, SalesRentals_flat_type, selected_flat, category, 'Rented') 
                               for selected_flat in SalesRentals_flat_types])
#Remove zeros
sold_percent = sold_percentages[sold_percentages != 0]
rented_percent = rented_percentages[rented_percentages != 0]
labels_sold = SalesRentals_flat_types[sold_percentages != 0]
labels_rented = SalesRentals_flat_types[rented_percentages != 0]
#specify the color
flat_type_colors = {
    '1-room flats': '#f5c7f5',  # Soft lavender
    '2-room flats': '#f0e68c',  # Pale yellow
    '3-room flats': '#a0d468',  # Light green
    '4-room flats': '#b3e0ff',  # Baby blue
    '5-room flats': '#d9b3ff',  # Lilac
    'Executive/Multi-Generation flat': '#ffb3b3',  # Light pink
    'Flats sold on short leases': '#ffd700',  # Golden yellow
    'Studio Apartments': '#ff9999',  # Light pink with a touch of red
}
#Plot the data
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 8))

ax1.pie(sold_percent,labels=labels_sold, autopct='%1.1f%%',colors=[flat_type_colors[label] for label in labels_sold], textprops={'fontsize': 10})
ax1.set_title('Flat Type Distribution: Sold', fontsize = 14, fontweight='bold')
ax2.pie(rented_percent,labels=labels_rented, autopct='%1.1f%%',colors=[flat_type_colors[label] for label in labels_sold], textprops={'fontsize': 10})
ax2.set_title('Flat Type Distribution: Rented', fontsize = 14, fontweight='bold')

plt.show()
```

**Sales** data shows **4-room flats** as the most popular, accounting for **almost half of all sales**. **5-room** and **3-room** flats follow, each representing **about one-fifth of sales**, while **2-room flats** comprise approximately 10%. Studio apartments, executive/multi-generation flats, and short-lease flats represent a small percentage of sales. 
In contrast, the **rental** market is dominated by **1-room flats (over half of all rentals)**, followed by **2-room flats (nearly half)**. **3-room** flats account for **6%**, **4-room** flats and **5-room** flats for a **negligible** amount.

```python
def unit_by_year(year_selected, no_of_units, unit_financial_year, category, category_selected):

    filter = (unit_financial_year == year_selected) & (category == category_selected)
    total_unit = np.sum(no_of_units[filter])
    
    return total_unit

line_sold = [unit_by_year(year_selected, no_of_units, SalesRentals_year, category, 'Sold') for year_selected in SalesRentals_years]
line_rented = [unit_by_year(year_selected, no_of_units, SalesRentals_year, category, 'Rented') for year_selected in SalesRentals_years]

# Plot the data
plt.figure(figsize=(10, 6))  # Set figure size

plt.plot(SalesRentals_years, line_sold, marker='o', linestyle='-', color='red', label="Sold")
plt.plot(SalesRentals_years, line_rented, marker='o', linestyle='--', color='blue', label="Rented")

# Add gridlines, labels, and title
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xlabel("Years", fontsize=12)
plt.ylabel("No of Units", fontsize=12)
plt.title("Units sold and rented per year", fontsize=14, fontweight='bold')

# Customize ticks and legend
plt.xticks(SalesRentals_years, rotation=45, fontsize=10)
plt.legend(fontsize=10, loc="upper left")

# Add annotations for better readability
for x, sold_unit, rented_unit in zip(SalesRentals_years, line_sold, line_rented):
    plt.text(x, sold_unit + 1000, f'{sold_unit:.0f}', ha='center', va='bottom', fontsize=8, color='red')
    plt.text(x, rented_unit - 1000, f'{rented_unit:.0f}', ha='center', va='top', fontsize=8, color='blue')
plt.show()
```

#### How do the sales and rental trends compare across different unit types?
Rental unit numbers remained relatively stable and low (below 5,000 units) throughout the observed period.
